```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
df = pd.read_csv ('Data_Raw_Water.csv')
```

```python
# Display the first few rows of the dataset
print("First few rows:")
print(df.head())

# Display the last few rows of the dataset
print("\nLast few rows:")
print(df.tail())

# Display the shape of the dataset (rows, columns)
print("\nShape of the dataset:")
print(df.shape)

# Display basic statistics of numeric columns
print("\nSummary statistics:")
print(df.describe())

# Display data types of each column
print("\nData types:")
print(df.dtypes)
```

```python
# convert data type of time to datetime
df[['time']] = df[['time']].apply(pd.to_datetime)
```

```python
#RAW WATER TURBIDITY TREND
plt.figure(figsize =(7, 4))
plt.plot(df[['time']], df[['turbidity']]) 
plt.legend(['TURBIDITY']);
plt.xticks(rotation=45);
#plt.ylim([0, 350]);
#plt.title('A graph of Turbidity against Time for the sensor-node at NYEWASCO-Raw Water Quality Section \n', size = 16)
plt.xlabel('Time (Date)', size = 14);
plt.ylabel('Turbidity (NTU)', size = 14);
#plt.tight_layout()
```

```python
#RAW WATER pH TREND
plt.figure(figsize =(7, 4))
plt.plot(df[['time']], df[['pH']]) 
plt.legend(['pH Value']);
plt.xticks(rotation=45);
#plt.ylim([7.2, 7.6]);
#plt.title('A graph of Water pH against Time for the sensor-node at NYEWASCO-Raw Water Quality Section \n', size = 16)
plt.xlabel('Time (Date)', size = 14);
plt.ylabel('Water pH', size = 14);
#plt.tight_layout()
```
